1\. Write a function that converts number representation, bin<->dec<->hex. (Clearly using the corresponding python built-in functions is not fair..)

```python
# Define all the digits till hexadecimal
DIGITS = "0123456789ABCDEF" 

def dec_to_base(n, base):
    if n == 0:
        return "0"

    # List to store digits
    result = []

    # General conversion algorithm 
    while n > 0:
        result.append(DIGITS[n % base]) # takes the digit corresponding to division rest
        n //= base # update n

    # returns a string where we take the elements of the list read backwords (reversed) without using separators ("")
    return "".join(reversed(result))

def base_to_dec(s, base):
    s = s.upper() # make sure that the string has all upper cases (for hex numbers to match DIGITS)

    value = 0
    for ch in s:
        i = DIGITS.index(ch) # returns the corresponding numeric index to the given char (ch=A –> 10)
        # value += ( i * base**( len(ch) - position of ch )
        value = value * base + i # accumulating the value from MSB without need of powers

    return value




# Direct functions using the previous ones
def dec_to_bin(number):
    return dec_to_base(number, 2)

def dec_to_hex(number):
    return dec_to_base(number, 16)

def bin_to_dec(number):
    return base_to_dec(number, 2)

def hex_to_dec(number):
    return base_to_dec(number, 16)

def bin_to_hex(number):
    return dec_to_hex( bin_to_dec(number) )

def hex_to_bin(number):
    return dec_to_bin( hex_to_dec(number) )


'''
# --- Universal converter: base <-> base ---
def convert(x: str, from_base: str, to_base: str) -> str:
    """
    Convert number x between {bin, dec, hex}.
    from_base and to_base ∈ {"bin", "dec", "hex"}.
    """
    base_map = {"bin": 2, "dec": 10, "hex": 16}
    fb = base_map[from_base]
    tb = base_map[to_base]

    # Step 1: convert to decimal (integer)
    if from_base == "dec":
        decimal = int(x)        # allowed because input already decimal
    else:
        decimal = base_to_dec(x, fb)

    # Step 2: decimal → target base
    if to_base == "dec":
        return str(decimal)
    else:
        return dec_to_base(decimal, tb)
'''



print( hex_to_dec("F31B") )
print( hex_to_bin("F31B") )

print( dec_to_hex(62235) )
print( dec_to_bin(62235) )


    
```

2\. Write a function that converts a 32 bit word into a single precision floating point (i.e. interprets the various bits as sign, mantissa and exponent)

```python
def float_interpretation(n):

    final_num = 0
    exp = 0
    mantissa = 1
    sign = 0
    
    for i, b in enumerate(n):
        # print(f"Bit {i} = {b}")
        if i == 0:
            if int(b) == 0:
                sign = 1
            else:
                sign = -1
            
        elif i <= 8 and i > 0:
            exp += int(b)*(2**(8-i))

        elif i > 8:
            mantissa += int(b)*2**(8-i)  

    final_num = sign * mantissa * 2**(exp-127)

    return final_num

x = "01001010010100101110101001010110"

print(float_interpretation(x))

```
