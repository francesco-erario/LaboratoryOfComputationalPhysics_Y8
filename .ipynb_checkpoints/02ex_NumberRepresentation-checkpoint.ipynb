1\. Write a function that converts number representation, bin<->dec<->hex. (Clearly using the corresponding python built-in functions is not fair..)

```python
# Define all the digits till hexadecimal
DIGITS = "0123456789ABCDEF" 

def dec_to_base(n, base):
    if n == 0:
        return "0"

    # List to store digits
    result = []

    # General conversion algorithm 
    while n > 0:
        result.append(DIGITS[n % base]) # takes the digit corresponding to division rest
        n //= base # update n

    # returns a string where we take the elements of the list read backwords (reversed) without using separators ("")
    return "".join(reversed(result))

def base_to_dec(s, base):
    s = s.upper() # make sure that the string has all upper cases (for hex numbers to match DIGITS)

    value = 0
    for ch in s:
        i = DIGITS.index(ch) # returns the corresponding numeric index to the given char (ch=A –> 10)
        # value += ( i * base**( len(ch) - position of ch )
        value = value * base + i # accumulating the value from MSB without need of powers

    return value




# Direct functions using the previous ones
def dec_to_bin(number):
    return dec_to_base(number, 2)

def dec_to_hex(number):
    return dec_to_base(number, 16)

def bin_to_dec(number):
    return base_to_dec(number, 2)

def hex_to_dec(number):
    return base_to_dec(number, 16)

def bin_to_hex(number):
    return dec_to_hex( bin_to_dec(number) )

def hex_to_bin(number):
    return dec_to_bin( hex_to_dec(number) )


'''
# --- Universal converter: base <-> base ---
def convert(x: str, from_base: str, to_base: str) -> str:
    """
    Convert number x between {bin, dec, hex}.
    from_base and to_base ∈ {"bin", "dec", "hex"}.
    """
    base_map = {"bin": 2, "dec": 10, "hex": 16}
    fb = base_map[from_base]
    tb = base_map[to_base]

    # Step 1: convert to decimal (integer)
    if from_base == "dec":
        decimal = int(x)        # allowed because input already decimal
    else:
        decimal = base_to_dec(x, fb)

    # Step 2: decimal → target base
    if to_base == "dec":
        return str(decimal)
    else:
        return dec_to_base(decimal, tb)
'''



print( hex_to_dec("F31B") )
print( hex_to_bin("F31B") )

print( dec_to_hex(62235) )
print( dec_to_bin(62235) )


    
```

2\. Write a function that converts a 32 bit word into a single precision floating point (i.e. interprets the various bits as sign, mantissa and exponent)

```python
def float_interpretation(n):

    final_num = 0
    exp = 0
    mantissa = 1
    sign = 0
    
    for i, b in enumerate(n):
        # print(f"Bit {i} = {b}")
        if i == 0:
            if int(b) == 0:
                sign = 1
            else:
                sign = -1
            
        elif i <= 8 and i > 0:
            exp += int(b)*(2**(8-i))

        elif i > 8:
            mantissa += int(b)*2**(8-i)  

    final_num = sign * mantissa * 2**(exp-127)

    return final_num

x = "01001010010100101110101001010110"

print(float_interpretation(x))

```

3\. Write a program to determine the underflow and overflow limits (within a factor of 2) for python on your computer. 

**Tips**: define two variables inizialized to 1 and halve/double them enough time to exceed the under/over-flow limits  

```python
i = 0
n = 1.0
prev_n = n

while True:
    prev_n = n
    n = n * 2
    if i % 100 == 0:
        print(f"Iteration {i} | n = {n:.3e}", flush=True)
    if n == float('inf'):
        print(f"Overflow at iteration {i}; max number doubling is {prev_n}", flush=True)
        break
    
    i += 1
```

4\. Write a program to determine the machine precision

**Tips**: define a new variable by adding a smaller and smaller value (proceeding similarly to prob. 2) to an original variable and check the point where the two are the same 

```python
var = 1.1
var_check = 1.
i = 0.1

while var != var_check :
    var = 1.0
    var += i
    print(f"var = {var} –> Precision: {i}")
    i /= 10
```

5\. Write a function that takes in input three parameters $a$, $b$ and $c$ and prints out the two solutions to the quadratic equation $ax^2+bx+c=0$ using the standard formula:
$$
x=\frac{-b\pm\sqrt{b^2-4ac}}{2a}
$$

(a) use the program to compute the solution for $a=0.001$, $b=1000$ and $c=0.001$

(b) re-express the standard solution formula by multiplying top and bottom by $-b\mp\sqrt{b^2-4ac}$ and again find the solution for $a=0.001$, $b=1000$ and $c=0.001$. How does it compare with what previously obtained? Why?

(c) write a function that compute the roots of a quadratic equation accurately in all cases

```python
import numpy as np
a = 0.001
b = 1000
c = 0.001

std_x = ((-b + np.sqrt(b**2 - 4*a*c))/(2*a) , (-b - np.sqrt(b**2 - 4*a*c))/(2*a))
print(f"Standard formula solution: {std_x}")

re_exp_x = ( (4*a*c)/(2*a*(-b - np.sqrt(b**2 - 4*a*c))) , (4*a*c)/(2*a*(-b + np.sqrt(b**2 - 4*a*c))))
print(f"Re-expressed formula solution: {re_exp_x}")

# I take the solutions with "-" sign so I don't do the subtraction between similar numbers but I do the sum
accurate_x = ( (4*a*c)/(2*a*(-b - np.sqrt(b**2 - 4*a*c))) , (-b - np.sqrt(b**2 - 4*a*c))/(2*a))
print(f"Working formula: {accurate_x}")
```

6\. Write a program that implements the function $f(x)=x(x−1)$

(a) Calculate the derivative of the function at the point $x = 1$ using the derivative definition:

$$
\frac{{\rm d}f}{{\rm d}x} = \lim_{\delta\to0} \frac{f(x+\delta)-f(x)}{\delta}
$$

with $\delta = 10^{−2}$. Calculate the true value of the same derivative analytically and compare with the answer your program gives. The two will not agree perfectly. Why not?

(b) Repeat the calculation for $\delta = 10^{−4}, 10^{−6}, 10^{−8}, 10^{−10}, 10^{−12}$ and $10^{−14}$. How does the accuracy scales with $\delta$?

```python
delta = [10e-2, 10e-4, 10e-6, 10e-8, 10e-10, 10e-12, 10e-14]
x_delta = [1+i for i in delta]
der = [ xi*(xi-1) / delta_i for xi, delta_i in zip(x_delta, delta)] # f(1) = 0

print(f"Calculated derivatives: \n {der}")

analytic_der = 1
print(f"Analytic derivative: {analytic_der}")

accuracy_change = [ d-1 for d in der ]
print(f"Accuracy change: \n {accuracy_change}")
```

7\. Consider the integral of the semicircle of radius 1:
$$
I=\int_{-1}^{1} \sqrt{(1-x^2)} {\rm d}x
$$
which it's known to be $I=\frac{\pi}{2}=1.57079632679...$.
Alternatively we can use the Riemann definition of the integral:
$$
I=\lim_{N\to\infty} \sum_{k=1}^{N} h y_k 
$$

with $h=2/N$ the width of each of the $N$ slices the domain is divided into and where
$y_k$ is the value of the function at the $k-$th slice.

(a) Write a programe to compute the integral with $N=100$. How does the result compare to the true value?

(b) How much can $N$ be increased if the computation needs to be run in less than a second? What is the gain in running it for 1 minute? 


```python
import math as m
import time

N = 1000000
h = 2/N
I = 0.

start = time.time()

for k in range(1,N+1):
    xk = h*k - 1
    I += h*m.sqrt(1 - xk**2)

true_I = m.pi / 2

elapsed = time.time() - start

print("Time:", elapsed)

print(f"True value of the integral: {true_I} | Computed value of the integral: {I}")
print(f"Comparison: {I -  true_I}")
    
```


#### Optimezed version without a loop:

```python
import numpy as np
import time

N = 100_000_000
h = 2 / N

start = time.time()
x = np.linspace(-1 + h, 1, N)
I = h * np.sum(np.sqrt(1 - x**2))
print("Time:", time.time() - start)

print("Error:", I - np.pi / 2)
```
