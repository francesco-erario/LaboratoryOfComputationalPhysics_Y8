1\. **Radioactive decay chain**

${\rm Tl}^{208}$ decays to ${\rm Pb}^{208}$ with a half-lieve of 3.052 minutes. Suppose to start with a sample of 1000 Thallium atoms and 0 of Lead atoms.

* Take steps in time of 1 second and at each time-step decide whether each Tl atom has decayed or not, accordingly to the probability $p(t)=1-2^{-t/\tau}$. Subtract the total number of Tl atoms that decayed at each step from the Tl sample and add them to the Lead one. Plot the evolution of the two sets as a function of time  
* Repeat the exercise by means of the inverse transform method: draw 1000 random numbers from the non-uniform probability distribution $p(t)=2^{-t/\tau}\frac{\ln 2}{\tau}$ to represent the times of decay of the 1000 Tl atoms. Make a plot showing the number of atoms that have not decayed as a function of time

```python
import numpy as np
import matplotlib.pyplot as plt

half_life = 3.052 * 60
tau = half_life / np.log(2)
p1s = 1 - np.pow(2., -1./tau)
print(f'Decay probability at every second is: {p1s}')

N_Tl0 = 1000
N_Pb0 = 0
N_Tl = []
N_Pb = []

time_span = 1000
N_Tl.append(N_Tl0)
N_Pb.append(N_Pb0)
for t in range(1,time_span):
    n = np.random.uniform(size=N_Tl[t-1]) # generate N_Tl random numbers uniformly distributed
    decayed_atoms = n[n<p1s].size # Number of decayed atoms based on probability
    N_Tl.append(N_Tl[t-1]-decayed_atoms)
    N_Pb.append(N_Pb[t-1]+decayed_atoms)

time = np.linspace(0,time_span,time_span)

plt.plot(time, N_Tl, '-', label='$Tl^{208}$')
plt.plot(time, N_Pb, '-', label='$Pb^{208}$')
plt.xlabel('time')
plt.legend()
plt.show()
```

```python
# Inverse transform method
z = np.random.uniform(size=N_Tl0)
t = -tau * np.log2(1-z)

N_survived = []
N_survived.append(N_Tl0)
for s in range(1,time_span):
    N_survived.append(t[t > s].size)

plt.plot(time, N_survived)
plt.show()
```

2\. **Rutherford Scattering**

The scattering angle $\theta$ of $\alpha$ particles hitting a positively charged nucleus of a Gold atom ($Z=79$) follows the rule:

$$
\tan{\frac{1}{2} \theta} = \frac{Z e^2} {2\pi \epsilon_0 E b}
$$

where $E=7.7$ MeV and $b$ beam is the impact parameter. The beam is represented by a 2D gaussian distribution with $\sigma=a_0/100$ for both coordinates ($a_0$ being the Bohr radius). Assume 1 million $\alpha$ particles are shot on the gold atom.

Computing the fraction of particles that "bounce back",i.e. those particle whose scattering angle is greater than $\pi/2$ (which set a condition on the impact parameter $b$)

```python
# Physical constants
Z = 79
e = 1.602176634e-19       # C
epsilon0 = 8.8541878128e-12
E = 7.7e6 * e             # MeV -> J
a0 = 5.29177210903e-11    # Bohr radius (m)

# Simulation params
N = 1_000_000
sigma = a0 / 100

# Critical impact parameter (condition on b for backwards scattering)
b_crit = Z * e**2 / (2 * np.pi * epsilon0 * E)

# Beam generation (2D gaussian)
x = np.random.normal(0, sigma, N)
y = np.random.normal(0, sigma, N)
# b parameter as distance from the center of the beam
b = np.sqrt(x**2 + y**2)

# Backwards scattering particle count
bounce = np.sum(b < b_crit)

fraction = bounce / N

print("Backwards scattering particle fraction:", fraction)
```

3\. **Monte Carlo integration: hit/miss vs mean value method**

Consider the function 

$$f(x) =\sin^2{\frac{1}{x(2-x)}}$$

* Compute the integral of $f(x)$ between 0 and 2 with the hit/miss method. Evaluate the error of your estimate
* Repeat the integral with the mean value method. Evaluate the error and compare it with the previous one

```python
def f(x):
    return (np.sin(1/(x*(2-x))))**2

x=np.linspace(0.001,1.999,1000)
plt.plot(x,f(x),'r-')

# Monte Carlo integration
N=100000
count=0
for i in range(N):
    x=2*np.random.random()
    y=np.random.random()
    if y<f(x): count+=1
I=2*count/N
print(I)

# The variance of k/N comes from a binomial distribution > Var(k/N) = p(1-p)/N
# Multiplying by A^2 (area of the rectangle) we obtain the estimate of the variance of the integral
# Since I  kA/N > Var(Ak/N)  A^2p(1-p)/N
# So the std is A sqrt( p(1-p)/N ) where p  k/N

error = 2 * np.sqrt( (count/N)*(1-(count/N))/N )
print('The estimate of the error is: ', error)
```

```python
# Mean value method
x_i = np.random.uniform(0,2, size=N)
I_mean = 2*f(x_i).mean()
print(I_mean)

error_mean = 2*f(x_i).std(ddof=1)/np.sqrt(N) # ddof=1 correzione di Bessel > errore sulla popolazione
print(error_mean)
```

4\. **Monte Carlo integration in high dimension**

* Start of by computing the area of a circle of unit radius, by integrating the function 

$$
f(x,y)=
\left\{
\begin{array}{ll}
      1 & x^2+y^2\le 1 \\
      0 & {\rm elsewhere}
\end{array} 
\right.
$$

* Generalize the result for a 10D sphere



```python
def g(x,y):
    if ((x**2 + y**2) <= 1):
        return 1
    else:
        return 0

n = 1000000
c = 0
for i in range(n):
    x = np.random.uniform(-1,1)
    y = np.random.uniform(-1,1)
    c += g(x,y)

Ig = 4 * c / n
print(Ig)


def G(r):
    if ( (r**2).sum() <= 1 ):
        return 1
    else:
        return 0

c10 = 0
for j in range(n):
    r = np.random.uniform(-1,1,size=10)
    c10 += G(r)

IG = 2**10 * c10 / n
print(IG)
print((np.pi**5) /120)
```

```python
# Faster version with mean method (heavier on RAM)
d = 10
n_mean = 1_000_000

x_mean = np.random.uniform(-1, 1, size=(n_mean, d))

r2 = np.sum(x_mean**2, axis=1)

f_mean = (r2 <= 1)

I_10D_mean = (2**d) * f_mean.mean()

print(I_10D_mean)
print(np.pi**5 / 120)
```

5\. **Monte Carlo integration with importance sampling** 

Calculate the value of the integral:

$$
I=\int_0^1 \frac{x^{-1/2}}{e^x+1} dx
$$

using the importance sampling method with $w(x)=1/\sqrt{x}$. You should get a result about 0.84

Steps to use this method:\
1.) Find a positive $w(x)$ to simplify $f(x)$\
2.) Calculate the normalization of $w(x)$ to make it a pdf \
3.) Generate random numbers distributed like $w(x)$ (normalized) \
4.) Use these numbers to calculate the mean of $f(x)/w(x)$ to get the value of the integral ($w(x)$ already normalized)

```python
def h(x): # f(x)/w(x) > with w(x) normalized [w(x)=1/2sqrt(x)]
    return (2/(np.exp(x)+1))

M = 1000000
zi = np.random.uniform(0,1,size=M)
Xi = zi**2 # X numbers are distributed like w(x) > inverse transform method

I_imp_samp = h(Xi).mean()
print(I_imp_samp)
```
