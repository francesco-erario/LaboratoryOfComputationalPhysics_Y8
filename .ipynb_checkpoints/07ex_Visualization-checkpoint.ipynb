1\. **Kernel Density Estimate**

Produce a KDE for a given distribution (by hand, not using seaborn!):

* Fill a numpy array, x,  of len(N) (with N=O(100)) with a variable normally distributed, with a given mean a standard deviation
* Fill an histogram in pyplot taking properly care about the aesthetic
   * use a meaningful number of bins
   * set a proper y axis label
   * set proper value of y axis major ticks labels (e.g. you want to display only integer labels)
   * display the histograms as data points with errors (the error being the poisson uncertainty)
* for every element of x, create a gaussian with the mean corresponding the element value and std as a parameter that can be tuned. The std default value should be:
$$ 1.06 * x.std() * x.size ^{-\frac{1}{5.}} $$
you can use the scipy function `stats.norm()` for that.
* In a separate plot (to be placed beside the original histogram), plot all the gaussian functions so obtained
* Sum (with np.sum()) all the gaussian functions and normalize the result such that the integral matches the integral of the original histogram. For that you could use the `scipy.integrate.trapz()` method


````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
from scipy.stats import norm
from scipy.integrate import trapezoid

mu, s, N = 5.0, 1.5, 200
x = np.random.normal(mu, s, N)

# Calculates values without plotting
counts, bin_edges = np.histogram(x, bins=15)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,6))

# First plot
ax1.set_ylabel('Counts')
errors = np.sqrt(counts)
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])
# ax.yaxis.set_major_locator(MultipleLocator())
ax1.errorbar(bin_centers, counts, yerr=errors, fmt='o', label='Poisson error')
ax1.hist(x, bins=15, histtype='step', label='Normal data')

# Second plot
std = 1.06*x.std()*(np.pow(x.size, (-1/5)))
x_plot = np.linspace(x.min() - 4*std, x.max() + 4*std, 1000)
gauss_sum = 0
for xi in x:
    pdf = norm.pdf(x_plot, loc=xi, scale=std)
    ax2.plot(x_plot, pdf, alpha=0.3)
    gauss_sum += pdf

ax1.legend()

integral = trapezoid(gauss_sum, x_plot)
print(integral)

````

2\. **Color-coded scatter plot**

Produce a scatter plot out of a dataset with two categories

* Write a function that generate a 2D datasets of 2 categories. Each category should distribute as a 2D gaussian with a given mean and std (clearly it is better to have different values means..)
* Display the dataset in a scatter plot marking the two categories with different marker colors.

An example is given below

You can try to make the procedure more general by allowing a given number $n\ge 2$ of categories

````python
! wget https://www.dropbox.com/s/u4y3k4kk5tc7j46/two_categories_scatter_plot.png
from IPython.display import Image
Image('two_categories_scatter_plot.png')
````

````python
import pandas as pd

# Generate a dataset with n features of size 'size'
def dataset_gen(n,size):
    dataset = {}
    for i in range(1,n+1):
        mean = np.random.randint(-5,5)
        std = np.random.uniform(0,3)
        feature = np.random.normal(mean, std, size=(size, 2))
        dataset[f'feature{i}_x'] = feature[:,0]
        dataset[f'feature{i}_y'] = feature[:,1]

    return pd.DataFrame(dataset)

df = dataset_gen(2,400)

fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['feature1_x'], df['feature1_y'], label='feature1', alpha=0.7)
ax.scatter(df['feature2_x'], df['feature2_y'], label='feature2', alpha=0.7)
plt.show()
````

3\. **Profile plot**

Produce a profile plot from a scatter plot.
* Download the following dataset and load it as a pandas dataframe:
```bash
wget https://www.dropbox.com/s/hgnvyj9abatk8g6/residuals_261.npy
```
Note that you should use the `np.load()` function to load the file as a numpy array, call the `.item()` method, and then pass it to the `pd.DataFrame()` constructor.
* Inspect the dataset, you'll find two variables (features)
* Clean the sample by selecting the entries (rows) with the variable "residual" in absolute value smaller than 2
* perform a linear regression of "residuals" versus "distances" using `scipy.stats.linregress()` 
* plot a seaborn jointplot of  "residuals" versus "distances", having seaborn performing a linear regression. The result of the regression should be displayed on the plot
* Fill 3 numpy arrays
  * x, serving as an array of bin centers for the "distance" variable. It should range from 0 to 20 with reasonable number of steps (bins)
  * y, the mean values of the "residuals", estimated in slices (bins) of "distance"
  * erry, the standard deviation of the  of the "residuals", estimated in slices (bins) of "distance"
* Plot the profile plot on top of the scatter plot

````python
from scipy.stats import linregress
import seaborn as sns

# Using .item() allows to get the content of the object "array(...)" instead of the object itsself
DF = pd.DataFrame(np.load('residuals_261.npy', allow_pickle=True).item())
print(DF.info())
DF = DF[ np.absolute(DF['residuals']) < 2 ]
reg = linregress(x=DF['distances'], y=DF['residuals'])
print(f'Results of the linear regression > slope = {reg.slope}, intercept = {reg.intercept}')

g = sns.jointplot(data=DF, x="distances", y="residuals", kind="reg", scatter_kws={'s': 15,'alpha': 0.3}, line_kws={'color': 'DarkBlue','lw': 2}, label='residuals')

# bin edges
bins = np.linspace(0,20,17)
# bin centers
x = 0.5 * (bins[:-1] + bins[1:]) # bins[:-1] goes from the first element to one before the end

y = []
erry = []
for i in range(len(bins) - 1): # len 21 is the number of edges, number of centers (bins) is actually 20
    mask = ( (DF['distances'] >= bins[i]) & (DF['distances'] < bins[i+1]) ) # & is a bitwise operator that works element by element ('and' does not work)
    y.append( DF.loc[mask, 'residuals'].mean() )
    erry.append( DF.loc[mask, 'residuals'].std() )

y = np.array(y)
erry = np.array(erry)

g.ax_joint.errorbar(x, y, yerr=erry, fmt='o', color='DarkRed', label='Profile') # ax_joint is an attribute of a JointGrid that returns a matplotlib object
g.ax_joint.legend()
````
