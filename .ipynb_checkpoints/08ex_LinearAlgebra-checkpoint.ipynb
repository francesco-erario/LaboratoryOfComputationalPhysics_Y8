1\. **PCA on 3D dataset**

* Generate a dataset with 3 features each with N entries (N being ${\cal O}(1000)$). With $N(\mu,\sigma)$ the normali distribution with mean $\mu$ and $\sigma$  standard deviation, generate the 3 variables $x_{1,2,3}$ such that:
    * $x_1$ is distributed as $N(0,1)$
    * $x_2$ is distributed as $x_1+N(0,3)$
    * $x_3$ is given by $2x_1+x_2$
* Find the eigenvectors and eigenvalues of the covariance matrix of the dataset
* Find the eigenvectors and eigenvalues using SVD. Check that the two procedures yield to same result
* What percent of the total dataset's variability is explained by the principal components? Given how the dataset was constructed, do these make sense? Reduce the dimensionality of the system so that at least 99% of the total variability is retained.
* Redefine the data in the basis yielded by the PCA procedure
* Plot the data points in the original and the new coordiantes as a set of scatter plots. Your final figure should have 2 rows of 3 plots each, where the columns show the (0,1), (0,2) and (1,2) proejctions.


```python
import numpy as np
import pandas as pd
from scipy import linalg as la

x1 = np.random.normal(0, 1, 2000)
x2 = x1 + np.random.normal(0, 3, 2000)
x3 = 2.0 * x1 + x2
df = pd.DataFrame({ 'x1':x1, 'x2':x2, 'x3':x3 })

cov = np.cov(df.T)
print(f'Covariance matrix is: \n{cov}')

eig, V = la.eigh(cov)
print(f'The eigenvalues of the covariance matrix are: {eig} \nthe V matrix is: \n{V}')

print('\nThe same happens with SVD method.')

Xc = df.values - df.values.mean(0) # Center data to apply PCA (columnwise)
U, s, Vt = la.svd(Xc)
print(s**2/(2000-1))
print(Vt.T)
```

Notice that one eigenvalue is $\approx 0$ because one column of the data is linearly dependent from the others. So one eigenvector will always be different because it refers to the eigenvector 0.

```python
Lambda = np.diag(eig)
print(f'Explained variability by PCA (first component): {Lambda[2,2]/Lambda.trace()}')
print(f'Explained variability by PCA (first and second): {(Lambda[1,1]+Lambda[2,2])/Lambda.trace()}')
```

```python
# Reduction of dataset
Xp = np.dot(V.T, Xc.T)
PCAdf = pd.DataFrame({'PC1':Xp[2,:], 'PC2':Xp[1,:]})
print(Xp)
PCAdf
```

```python
import matplotlib.pyplot as plt

fig, ((px12,px13,px23),(pca12,pca13,pca23)) = plt.subplots(2,3, figsize=(15,6))
px12.scatter(Xc[:,0], Xc[:,1], alpha=0.7, label='x2 vs x1')
px13.scatter(Xc[:,0], Xc[:,2], alpha=0.7, label='x3 vs x1')
px23.scatter(Xc[:,1], Xc[:,2], alpha=0.7, label='x3 vs x2')
pca12.scatter(Xp[0,:], Xp[1,:], alpha=0.7, label='pca2 vs pca1')
pca13.scatter(Xp[0,:], Xp[2,:], alpha=0.7, label='pca3 vs pca1')
pca23.scatter(Xp[1,:], Xp[2,:], alpha=0.7, label='pca3 vs pca2')
px12.legend()
px13.legend()
px23.legend()
pca12.legend()
pca13.legend()
pca23.legend()
plt.show()
```
