2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

````python
import numpy.random as npr

random_matrix = npr.rand(5,5)
print(random_matrix)

with open("05_data/data.txt", mode="w") as df:
    for line in random_matrix:
        df.write('\t'.join(map(str,line)) + '\n')
````

### Alternative with numpy.savetxt() 

```python
import numpy as np
np.savetxt('data.txt', random_matrix, fmt='%.8f') # omitting fmt it saves numbers in scientific notation with 18 decimals
```

3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

````python
data = []
with open("05_data/data.txt", mode="r") as df:
    for line in df:
        l = []
        for c in line.split('\t'):
            l.append(float(c))
        data.append(l)

with open("05_data/data.csv", mode="w") as csvf:
    for line in data:
        csvf.write(','.join(map(str,line)) + '\n')
````

Instead of the second cycle one can write code using map():
```python

```
Usually one writes:
```python
line.strip().split()
```
Because there can be white spaces sometimes and if they happen to be inside float() conversion, for example, the compiler gives an error; strip() function prevents this erasing space chars eventually.

Very easy and rapid version:

4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line corresponds to a credit card number.
Each character is composed by 6 bits (even the space) and the last 4 bits are just a padding

**hint**: use the `chr()` function to convert a number to a char

````python
with open("credit_card.dat", mode='r') as ccd:
    for line in ccd:
        raw_number = [int(line[i:i+6],2) for i in range(0,len(line),6)] # converts from bin to dec every chunk of 6 bits in the line
        del raw_number[-1] # erase last element
        cc_number = ''.join(map(str,map(chr,raw_number)))
        print(cc_number)
````

5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.

````python
import json
with open("user_data.json", mode='r') as jf:
    users = json.load(jf)

# Parse American Express users
amex_users = [ user for user in users if user.get("CreditCardType") == "American Express" ]
keys = [ key for key in amex_users[0] ] # Take the first dictionary and make a list of keys

with open("american_express_users.csv", mode='w') as axuf:
    axuf.write(','.join(map(str,keys)) + '\n')
    for user in amex_users:
        # Take value from each dictionary (user) and write them in the file
        axuf.write(','.join(map(str,[user[key] for key in keys])) + '\n')
````

6\. Load the file from this url: [https://www.dropbox.com/s/7u3lm737ogbqsg8/mushrooms_categorized.csv?dl=1](https://www.dropbox.com/s/7u3lm737ogbqsg8/mushrooms_categorized.csv?dl=1) with Pandas.

````python
import pandas as pd
df = pd.read_csv('https://www.dropbox.com/s/7u3lm737ogbqsg8/mushrooms_categorized.csv?dl=1')

print("First rows: \n", df.head(5)) # Preview of the first 5 rows
print("Last rows: \n", df.tail(5)) # Preview of the last 5 rows

print('\n')
print(df.info()) # Provides a summery of the features of the dataset
print('\n')
print(df.describe()) # Provides a simple analysis of the data
print('\n')
print(df.shape) # Provides number of rows and columns

%matplotlib inline
import matplotlib.pyplot as plt

df['class'].value_counts().plot(kind='bar') # Categoric histogram â€”> there are only 2 classes
plt.xlabel('class')
plt.ylabel('count')
plt.show()
````
