1\. Find the row, column and overall means for the following matrix:

```python
m = np.arange(12).reshape((3,4))
```

````python
import numpy as np

m = np.arange(12).reshape((3,4))
print(m)
print("Mean of the rows:")
print("Row 0 –> ", m[0,:].mean())
print("Row 1 –> ", m[1,:].mean())
print("Row 2 –> ", m[2,:].mean())

print("Mean of the columns:")
print("Col 0 –> ", m[:,0].mean())
print("Col 1 –> ", m[:,1].mean())
print("Col 2 –> ", m[:,2].mean())
print("Col 3 –> ", m[:,3].mean())

print("Overall mean: ", m.mean())
````

2\. Find the outer product of the following two vecotrs

```python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])
```

Do this in the following ways:

   * Using the function outer in numpy
   * Using a nested for loop or list comprehension
   * Using numpy broadcasting operations


````python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])

# tensor product <–> outer product
out1 = np.outer(u, v)
print("np.outer method: \n", out1)

out2 = np.array([[i*j for j in v] for i in u])
print("Nested list method: \n", out2)

u_b = np.tile(u, (4,1)).T
v_b = np.tile(v, (4,1))
out3 = (u_b * v_b)
print("Broadcast method: \n", out3)
````

4\. Use np.linspace to create an array of 100 numbers between 0 and 2π (inclusive).

  * Extract every 10th element using slice notation
  * Reverse the array using slice notation
  * Extract elements where the absolute difference between the sine and cosine functions evaluated at that element is less than 0.1
  * Make a plot showing the sin and cos functions and indicate where they are close

````python
x = np.linspace(0,np.pi,100)
print("x array: \n", x)
print("x sliced every 10 elements: \n", x[::10])

y = x[::-1]
print("Reversed x: \n", y)

mask = ( np.absolute(np.cos(x)-np.sin(x)) < 0.1 )
print("Extracted values: \n", x[mask])

import matplotlib.pyplot as plt

sinx = np.sin(x)
cosx = np.cos(x)

plt.figure(figsize=(8, 5))

# sin and cos curves
plt.plot(x, sinx, label='sin(x)')
plt.plot(x, cosx, label='cos(x)')

# points where they are near each other
plt.scatter(x[mask], sinx[mask], color='red', zorder=3,
            label='|sin(x) - cos(x)| < 0.1')

plt.xlabel('x')
plt.ylabel('value')
plt.legend()
plt.title('sin(x) and cos(x) with close points highlighted')
plt.grid(True)

plt.show()
````

7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Constract a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
import time

N = 100
numbers = np.arange(N)
mask = [] # Final mask to identify prime numbers

start = time.perf_counter()

for n in numbers:  # Selects every number in the array
    truth_list = [] # Empty list to store new bool values at every number
    for i in range(2,N):
        if i != n: # Skip the current number that would give a True
            truth_list.append( (n % i == 0) ) # Store bool values in the list for every integer
    mask.append(any(truth_list)) # Set True if n has a perfect division and store the result (any() works like np.any() in this case)

mask[1] = True # Excludes 1 from prime numbers
print(f"Total time: {time.perf_counter()-start} seconds")
print(mask)
print("\n")
primes = numbers[np.logical_not(mask)]
print(f"Prime numbers in range {N} are: \n {primes}")
````

For N = 100 we have 0.002501582999684615 seconds
For N = 1000 we have 0.18152754200127674 seconds
For N = 10000 we have 12.615258415997232 seconds

–> It scales like $O(N^2)$

````python
# My version with np array
numbers_a = np.arange(N)
mask_a = np.zeros(N, bool) # Creates False array

divisors = np.arange(2, N) # Creates array of divisors from 2 to N

start = time.perf_counter()

for n in numbers_a: # Loops through the numbers
    if n < 2:
        mask_a[n] = True # Sets True elements of the mask less than 2 and continues the loop
        continue

    # Calculates rests in one go (sets True if n is perfectly divisible for a number in divisors –> numbers that are multiples)
    is_divisible = (n % divisors == 0)

    # Sets False the case n divided by itself that doesn't count
    is_divisible[n-2] = False

    # If there's a True in is_divisible (n is a multiple of something) the mask is set to True
    mask_a[n] = np.any(is_divisible)

print(f"Total time: {time.perf_counter()-start} seconds")
print(f"Prime numbers in range {N} are: \n {numbers_a[np.logical_not(mask_a)]}")
````

````python
# Sieve of Eratosthenes
all_num = np.arange(N)
is_prime = np.ones(N, bool) # create an array with all True
is_prime[:2] = False # set first 2 values (0 and 1) False

start = time.perf_counter()

for p in range(2, int(np.sqrt(N)) + 1): # it is enough that p goes until sqrt(N)+1 
    if is_prime[p]:
        is_prime[p*p:N:p] = False # Sets False (not prime) every number multiple of p (that is marked as prime - True) starting from p^2

print(f"Total time: {time.perf_counter()-start} seconds")
print(f"Prime numbers in range {N} are: \n {all_num[is_prime]}")
````
