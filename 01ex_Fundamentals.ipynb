3\. Write a `decorator` hello that makes every wrapped function print “Hello!”, i.e. something like:

```python
@hello
def square(x):
    return x*x
```



````python
def hello(func):
    def wrapped(x):
        print("Hello!")
        print(func(x))
    return wrapped

@hello
def square(x):
    return x*x
````

4\. Write the factorial function so that it a) does and b) does not use recursion.

````python
def fact_a(n):
    if n == 0:
        return 1
        
    else:
        return n * fact_a(n-1)

print(fact_a(5))
````

````python
def fact_b(n):
    x = 1
    while n>0:
        x *= n
        n -= 1
    return x

print(fact_b(5))
````

5\. Use HOFs (zip in particular) to compute the weight of a circle, a disk and a sphere, assuming different radii and different densities:

```python
densities = {"Al":[0.5,1,2],"Fe":[3,4,5],"Pb": [15,20,30]}
radii = [1,2,3]
```

where the entries of the dictionary's values are the linear, superficial and volumetric densities of the materials respectively.

In particular define a list of three lambda functions using a comprehension that computes the circumference, the area and the volume for a given radius.


````python
import numpy as np

densities = {"Al":[0.5,1,2],"Fe":[3,4,5],"Pb": [15,20,30]}
radii = [1,2,3]

geo_func = [lambda r: 2*np.pi*r, lambda r: np.pi*r**2, lambda r: (4/3)*np.pi*r**3]

# prova = [f(1.) for f in geo_func]
# print(prova)

weight = { material: { 
                    r: [f(r)*d for f,d in zip(geo_func, densities[material])] for r in radii
                     } 
          for material in densities 
         }

print(weight)
````

````python
# Parent class
class Dog:

    # Class attribute
    species = 'mammal'

    # Initializer / Instance attributes
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.is_hungry = True

    # instance method
    def description(self):
        return "{} is {} years old".format(self.name, self.age)

    # instance method
    def speak(self, sound):
        return "{} says {}".format(self.name, sound)

    def eat(self):
        if self.is_hungry == True:
            self.is_hungry = False

Pippo = Dog("Pippo", 3)
print(Pippo.is_hungry)

Pippo.eat()
print(Pippo.is_hungry)


# Create dog objects
dog1 = Dog("Tom", 6)
dog2 = Dog("Fletcher", 7)
dog3 = Dog("Larry", 9)

dogs = [dog1, dog2, dog3]

# Feed each dog
for dog in dogs:
    dog.eat()

# Check hunger status
# All checks if all elements are true
all_hungry = all(dog.is_hungry for dog in dogs)
all_not_hungry = all(not dog.is_hungry for dog in dogs)

# Output
print(f"I have {len(dogs)} dogs.", end="  ")
print(f"{dog1.name} is {dog1.age}. ", end="")
print(f"{dog2.name} is {dog2.age}. ", end="")
print(f"{dog3.name} is {dog3.age}. ", end="")
print("And they're all mammals, of course.", end="  ")

if all_hungry:
    print("My dogs are hungry.")
elif all_not_hungry:
    print("My dogs are not hungry.")
else:
    print("Some of my dogs are hungry.")
````
